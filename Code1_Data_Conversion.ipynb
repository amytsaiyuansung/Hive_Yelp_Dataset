```python
# Checkin

import json
import pandas as pd
import urllib
from urllib.request import urlopen

with open("checkin.txt", 'rb') as f:
    content = f.readlines()
    
json_data = [json.loads(line) for line in content]
checkin = pd.DataFrame(json_data)
checkin.to_csv('checkin.txt',sep='|',na_rep='NA',header=True, index=False)
checkin.head()
```

```python
# Business

import json
import pandas as pd
import urllib
from urllib.request import urlopen

with open("business.txt", 'rb') as f:
    content = f.readlines()
    
json_data = [json.loads(line) for line in content]

business = pd.DataFrame(json_data)
business = business.drop(['address','hours','latitude','longitude','postal_code'],axis=1)

business.to_csv('business.txt',sep='|',na_rep='NA',header=True, index=False)

business.head()
```
